```python
a = np.array([[2, -1, 3, 0], [-1, 4, -1, 10], [0, 10, 15, 50], [20, 10, 1, 40]])

b = np.array ([[4, 1, 0, 4], [5, -3, 1, 2], [-9, 2, -1, 5]])

rangoa = np.linalg.matrix_rank(a)

rangob = np.linalg.matrix_rank(b)

print('Rango de la matriz A: ', rangoa ,'Al ser el rango igual al tamao de la matriz, se tiene una unica solucin \n')
print('Rango de la matriz B: ', rangob , 'Al ser el rango diferente al tamao de la matriz, se tienen infinitas soluciones \n')

print('Ahora bien, podemos saber la nulidad para saber el numero de parametros que hay en la solucin: ', 4-rangob)

```

```python
A = np.array([[-5,0], [0,3]])
evals, evecs = np.linalg.eig(A)

print ('Los autovalores de A son: \n',evals)
print ('Los autovectores de A son: \n', evecs)

B = np.array([[-1, 2], [5,2]])
evals1, evecs1 = np.linalg.eig(B)

print ('Los autovalores de B son: \n ',evals1)
print ('Los autovectores de B son: \n', evecs1)
```
